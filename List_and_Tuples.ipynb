List: Built in data types that stores set of Values.
It can store elements of different types, Integer,float,strings etc.
Lists are mutable in Python bur strings are immutable.
Indexing is allowed in List.

```python
Student=["Ashmin", 98, 17, "Kiratichhap"]
print(Student[0])
Student[0]= "Ashmin Kunwar"
print(Student[0])
print(Student)
```

List Slicing: Similar to string. list_name[starting_index:ending_index]

```python
marks=[77,89,67,88,54]
print(marks[1:4])
```

List Methods: Append,sort,reverse,insert

```python
List=[5,7,8]
List.append(4)
print(List)
#Sorting: Ascending and Descending Order
#print(List.sort())#Gives none as a output (reverse=True) returns output in Descending Order.
#print(List)
List.reverse()
print(List)
List.insert(3,9)
print(List)
```

```python
#list.remove
#list.pop, removes element of given index.
age=[12,45,23,12]
age.remove(45)
print(age)
age.pop(1)
print(age)
```

Tuples: Built in data type that lets us create immutable sequences of values.
Stored inside parenthesis.

```python
tup=(5,7,8,9)
print(tup[0])
print(type(tup))
name=("Ashmin","Kunwar")
print(name)
print(tup[1:3])
```

```python
num=(4,5,7,9,0)
print(num.index(9))
print(num.count(5))
```
