Strings: It is a data type that stores a sequence of characters.

```python
str1="My name is Ashmin Kunwar.\n I am an agriculture graduate."
print(str1)
# Escape sequence characters
#\n = next line
```

Concatenation: Adding two strings

```python
str1="Ashmin"
str2="Kunwar"
str3=str1+" "+str2
print(str3)
len(str1)
```

Indexing: position number of any character. Like roll numbers of students. Indexing starts from 0. It helps to access characters. code style: str[3]. We cannot change characters with the help of string.

```python
name="Ashmin Kunwar"
print(name[8])
```

Slicing: Accessing parts of a string. Code style:str[starting_idx:ending_idx]
Negative Index: Starting from -1 counting backward.

```python
name="railagang"
print(name[5:])
print(name[-6:-1])
print(name.endswith("f"))
print(name.count("a")) #count the occurence of "a"
print(name.capitalize())
name=name.capitalize()
print(name.find("l"))# finds the index of character
print(name.replace("l","I"))# replace the character
print(name.find("gang"))# finds the word
```

Conditional Statements: if-elif-else (Syntax-the rule of programming)
if (condition):
  statement1
elif (condition):
  statement2
else (condition):
  statement3

```python
age=int(input("Enter your age:"))
if age>=18:
  print("You are eligible to vote and can apply for liscense")
else :
  print("You are not eligible to vote and apply for liscense")
```

```python
light= input("Enter the colour of light:")
if light=="red":
  print("Stop")
elif light=="yellow":
  print("Get Ready")
elif light=="green":
  print("Go")
  #Indentation: it means proper spacing
```

```python
marks=int(input("Enter your marks:"))
if marks>=90:
  print("Grade A")
elif marks>=80 and marks<90:
  print("Grade B")
elif marks>=70 and marks<80:
  print("Grade C")
elif marks>=60 and marks<70:
  print("Grade D")
else:
  print(" Fail")
```

Nesting: Writing one statement inside another statement, which is valid in Python.

```python
age=int(input("Enter your age:"))
if age>=18:
  if age>=80:
    print("You cannot drive")
  else:
    print("You can drive")
else:
  print("You cannot drive")
```
