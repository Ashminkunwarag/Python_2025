Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import resample
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
import pickle
```

Data Collection and Processing

```python
# Load the dataset to a pandas dataframe
data=pd.read_csv("/Rainfall.csv")
print(type(data))
data.shape
data.head()# To know first five rows of thd data.
data.tail()
data["day"].unique()
print("Data Info:")
data.info() # To know about missing values and data types of various columns
```

```python
# Converting the yes and no to 1 and 0 respectively
data["rainfall"]=data["rainfall"].map({"yes":1,"no":0})
```

**Data Preprocessing**

```python
# Drop highly correlated column
data=data.drop(columns=['maxtemp', 'temparature', 'mintemp'])
```

```python
data.head()
```

```python
print(data["rainfall"].value_counts())
```

```python
# Separate majority and minority class
df_majority = data[data.rainfall==1]
df_minority = data[data.rainfall==0]
```

```python
print(df_majority.shape)
print(df_minority.shape)
```

```python
# Downsample majority class to match minority count
df_majority_downsampled = resample(df_majority,
                                    replace=False,
                                    n_samples=len(df_minority),
                                    random_state=42)
```

```python
df_majority_downsampled.shape
```

```python
df_downsampled=pd.concat([df_majority_downsampled,df_minority])
```

```python
df_downsampled.shape
```

```python
df_downsampled.head()
```

```python
# Shuffle the final dataframe
df_downsampled=df_downsampled.sample(frac=1,random_state=42).reset_index(drop=True)
```

```python
df_downsampled.head()
```

```python
df_downsampled["rainfall"].value_counts()
```

```python
# Split features and target as X and Y
x=df_downsampled.drop(columns=["rainfall"])
y=df_downsampled["rainfall"]
```

```python
print(x)
```

```python
print(y)
```

```python
# Splitting the data into training data and test data
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

**Model Training**

```python
rf_model = RandomForestClassifier(random_state=42)

param_grid_rf= {
    "n_estimators": [50,100,200],
    "max_feature": ["sqrt","log2"],
    "max_depth": [None,10,20,30],
    "min_samples_split": [2,5,10],
    "min_samples_leaf": [1,2,4]
}
```

```python
# Hypertuning using Gridsearch CV
# Corrected parameter name from 'max_feature' to 'max_features'
rf_model = RandomForestClassifier(random_state=42)

param_grid_rf= {
    "n_estimators": [50,100,200],
    "max_features": ["sqrt","log2"],
    "max_depth": [None,10,20,30],
    "min_samples_split": [2,5,10],
    "min_samples_leaf": [1,2,4]
}

grid_search_rf=GridSearchCV(estimator=rf_model,param_grid=param_grid_rf,cv=5,n_jobs=-1,verbose=2)
grid_search_rf.fit(x_train,y_train)
```

```python
best_rf_model=grid_search_rf.best_estimator_
print("Best Parameter for Random Forest:",grid_search_rf.best_params_)
```

**Model Evaluation**

```python
cv_scores=cross_val_score(best_rf_model,x_train,y_train,cv=5)
print("Cross Validation Scores:",cv_scores)
print("Mean Cross Validation Score:",np.mean(cv_scores))
```

```python
# test set performance
y_pred=best_rf_model.predict(x_test)
print("Test of accuracy:",accuracy_score(y_test,y_pred))
print("Test of confusion matrix:\n",confusion_matrix(y_test,y_pred))
print("Test of classification report:\n",classification_report(y_test,y_pred))
```
