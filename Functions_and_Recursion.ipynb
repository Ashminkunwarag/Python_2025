Functions: A block of statements that performs a specific task.

```python
# def func_name(param1,param2):
# some work
# return value
# func_name(arg1,arg2,...)  # Function Calling
def calc_sum(a,b):
  sum= a+b
  return sum
calc_sum(10,20)
```

```python
#Function definition
def calc_sum(a,b): # Parameters a and b
  return a+b

sum = calc_sum(10,20) # Function call; arguments
print(sum)
```

```python

def calc_avg(a,b,c):
  sum= a+b+c
  avg= sum/3
  return avg

calc_avg(10,20,30)
```

User-defined function: Built by a user.
Default Parameter: default value to parameter, so that program can run when there is no any argument.


```python
def calc_sum(a=4,b=5):
  sum= a+b
  return sum
calc_sum()
```

```python
# Practice
def calc_fact(n):
  fact = 1
  for i in range(1, n+1):
    fact *= i
  print (fact)
calc_fact(5)
```

```python
# USD to NPR
def converter(usd_val):
  npr_val = usd_val * 130
  print(usd_val, "USD =", npr_val, "NPR")
converter(100)
```

Recursion: When a function calls itself repeatedly.


```python
# Recursive Function
def show(n):
  if (n ==0): #Base case
    return
  print(n)
  show(n-1)
show(5)
```

File I/O in Python:
Doing operations in file. For example: read and write data.
Text files and Binary  files. All files are bits.
We have to open a file before reading or writing.
Syntax: f=open("filename", "mode")
data = f.read()
f.close()
"r" for reading, "w" for writing and truncating, "x" for creating new file and open for writing, "t" for text mode.

```python
# Creating 'my_file.txt' for demonstration using with open
with open("my_file.txt", "w") as f:
  f.write("This is a test line.\n")
  f.write("This is another test line.")
print("my_file.txt created with sample content.")
```

```python
# Appending text to 'my_file.txt'
with open("my_file.txt", "a") as f:
  f.write("\nThis is an appended line.")
print("Appended 'This is an appended line.' to my_file.txt")

# Read and display the updated content of 'my_file.txt'
with open("my_file.txt", "r") as f:
  updated_content = f.read()
print("\nUpdated content of my_file.txt:")
print(updated_content)
# more can be learnt from stackoverflow.com
```
