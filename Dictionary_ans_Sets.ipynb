Dictionary in Python:
Used to store data values in key:value pairs.
They are unordered, mutable(Changeable), and dont allow duplicate keys.

```python
info={
    "key": "value",
    "name": "Ashmin",
    "age": 26,
    "University": "Agriculture and Forestry University",
    "CGPA": 3.27,
}
print(info)
info["name"] = "Ashmin_Kunwar"
print(info)
info["Job"] = "Agriculture Officer"
print(info)
```

```python
null_dict = {}
null_dict["key"] = "value"
print(null_dict)
# We can add key values in null dictionary.
```

```python
# Nested Dictionaries
Student = {
    "name": "Ashmin",
    "Class": 10,
    "Subjects": {
        "Math": 91,
        "Science": 82,
        "English": 85,
        "Nepali": 75,
        "Social": 74,
        "HPE": 89,
        "Additional Math": 78,
        "Computer": 91,
    },
    "Percentage": 83.13,
}
print(Student)
print(Student["Subjects"])
print(Student["Subjects"]["Math"])
# Commas are necessary after  key value pairs.
```

```python
#Methods
print(Student.keys())
print(Student.values())
print(Student.items()) # Returns all key, value pairs as tuples
print(list(Student.keys()))
print(list(Student.values()))
print(list(Student.items()))
print(Student.get("Subjects"))
Student.update({"Address": "Kiratichhap"})
print(Student)
```

Set in Python:
Set is the collection of the unordered item.
Each element in the set must be unique and immutable.
Example: nums = {1,2,3,4}
nullset= set(), Syntax for null set

```python
collection = {1,5,7,"Hello",3.27, 7.5,"Ashmin",5}
print(collection)
#Set Methods
collection.add(26) # adds the element
print(collection)
collection.remove("Hello") # removes the element
print(collection)
collection.pop() # removes any random values
print(collection)
collection.clear() # empties the set
print(collection)
```

```python
# Union and Intersection
set1 = {1,2,3,4,5}
set2 = {4,5,6,7,8}
print(set1.union(set2))
print(set1.intersection(set2))
```
