Loops: Used to repeat instructions.
While loop and For loop in Python.
while condition:

```python
count=1
while count <= 5:
    print("Hello")
    count=count+1
```

```python
i=1 # Iterator
while i <=10:
  print (i)
  i += 1
print ("Loop Ended")
# There  is stopping condition in loop. Infinite loops are dangerous in programming and it is unnecessary.
```

```python
i=10
while i >=1: # Stopping condition
  print (i)
  i -= 1
print ("Loop Ended")
```

```python
# Multiplication Table
n=6
i=1
while i <=10: # Stopping condition
  print (n*i)
  i += 1
```

```python
#Traverse
num = [1,4,9,16,25,36,49,64]
idx = 0
while idx < len(num):
  print (num[idx])
  idx += 1
```

```python
# Search in tuple
num = (1,4,9,16,25,36,49,64,81,100)
x=36
i = 0 # Initialization
while i < len(num):
  if num[i] == x:
    print ("Found at index", i)
  i += 1
```

Break and Continue
Break: Used to terminate loop when encountered.
Continue: terminate execution in current iteration and continue execution of the loop in next iteration.

For Loop:
Used for Sequential traversal. For traversing list,string,tuples etc.


```python
list = [1,3,5,6,7,8]
for value in list:
  print(value)
else:
  print("End")
```

```python
#Searching Index
nums = (1,4,9,16,25,36,49,64,81,100)
x = 49

i = 0
for value in nums:
  if value == x:
    print ("Found at index", i)
    break
  i += 1
else:
  print ("Not Found")

```

Range():
Range function returns sequence of numbers, starting from 0 by default and increament by 1 (default), and stops before a specified number.

```python
# range(start?,stop,step?)
for i in range(1,10,2):
  print(i)
```
